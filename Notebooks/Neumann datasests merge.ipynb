# Create datasets

```python
import shutil

out_path = r"C:\Users\Martin\Desktop\datasets"
os.makedirs(out_path, exist_ok=True)

for class_name, class_items_list in class_items_directories.items():
    class_name_path = os.path.join(out_path, class_name)
    os.makedirs(class_name_path, exist_ok=True)
    for class_item_folder in class_items_list:
        class_item_folder_path = dir_path + os.sep + class_item_folder
        if os.path.isdir(class_item_folder_path):
        
            for image_item in os.listdir(class_item_folder_path):
                image_item_path = class_item_folder_path + os.sep + image_item
                if os.path.isfile(image_item_path):
                    shutil.copy(src=image_item_path, dst=class_name_path + os.sep + image_item)
```
